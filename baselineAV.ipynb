### Model 2 - XGB 

```python
## check missing values

misCols = train.isnull().sum(axis=0)/train.shape[0]
```

```python
## drop columns with 90% missing values
toremove = []

for k, v in misCols.iteritems():
    if v >= 0.9:
        toremove.append(k)
```

```python
train.drop(toremove, inplace=True, axis=1)
test.drop(toremove, inplace=True, axis=1)
```

```python
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for c in train.columns:
    if train[c].dtype == 'object' and c not in ['CUSTOMER_ID','RESPONDERS']:
        #print(c)
        lbl = LabelEncoder()
        lbl.fit(list(train[c].values) + list(test[c].values))
        train[c] = lbl.transform(list(train[c].values))
        test[c] = lbl.transform(list(test[c].values))
```

```python
train['RESPONDERS'] = train['RESPONDERS'].replace(to_replace = {'Y':1, 'N':0})
```

```python
## using xgb
import xgboost as xgb
```

```python
target = train['RESPONDERS']
```

```python
X_train = train.drop(['CUSTOMER_ID','RESPONDERS'], axis=1)
X_test = test.drop(['CUSTOMER_ID'], axis=1)
```

```python
dtrain = xgb.DMatrix(X_train, label=target, missing=np.nan)
dtest = xgb.DMatrix(X_test, missing=np.nan)
```

```python
params = {"objective": "binary:logistic","booster": "gbtree", "nthread": 4,
                "eta": 0.08, "max_depth": 6, "subsample": 0.9, "colsample_bytree": 0.7,
                "min_child_weight": 1,
                "seed": 2016, "tree_method": "exact"}
```

```python
xgb_cv = xgb.cv(params, dtrain, num_boost_round=1000, nfold=10, stratified=True, metrics = 'auc', \
                maximize=True, early_stopping_rounds=40, verbose_eval=20)
```

```python
best_round = np.argmax(xgb_cv['test-auc-mean'])
```

```python
xgb1 = xgb.train(params, dtrain, num_boost_round=best_round)
```

```python
pred1 = xgb1.predict(dtest)
```

```python
sub['RESPONDERS'] = pred1
```

```python
sub.to_csv('submissions/xgb_baseline.csv', index=False) ##0.5643
```
